```python
import pandas as pd

df = pd.read_csv('data/r4n/dataset_r4n_c20_with_cid.csv')
df
```

```python
# 检查数据框结构和SMILES列
df.info()
df.head()
```

```python
import pubchempy as pcp
import time
from tqdm import tqdm

def get_xlogp_from_smiles(smiles):
    """通过SMILES从PubChem查询XLogP值"""
    try:
        compounds = pcp.get_compounds(smiles, namespace='smiles')
        if compounds:
            comp = compounds[0]
            # 获取XLogP属性
            xlogp = comp.xlogp
            return xlogp
        return None
    except Exception as e:
        print(f"Error for SMILES {smiles}: {e}")
        return None

# 创建新列存储从PubChem查询的XLogP
df['xlogp_pubchem'] = None

# 对每个SMILES查询XLogP（带进度条）
for idx in tqdm(df.index, desc="Querying PubChem"):
    smiles = df.loc[idx, 'SMILES']
    xlogp_value = get_xlogp_from_smiles(smiles)
    df.loc[idx, 'xlogp_pubchem'] = xlogp_value
    # 添加小延迟避免请求过快
    time.sleep(0.2)

print(f"\n完成查询！")
print(f"成功获取 {df['xlogp_pubchem'].notna().sum()} 个XLogP值")
print(f"缺失 {df['xlogp_pubchem'].isna().sum()} 个XLogP值")
```

```python
df
```

```python
# 重新查询缺失的XLogP值
missing_indices = df[df['xlogp_pubchem'].isna()].index
for idx in tqdm(missing_indices, desc="Re-querying missing XLogP"):
    smiles = df.loc[idx, 'SMILES']
    xlogp_value = get_xlogp_from_smiles(smiles)
    df.loc[idx, 'xlogp_pubchem'] = xlogp_value
    time.sleep(0.2)

print(f"\n重新查询完成！")
print(f"成功获取 {df['xlogp_pubchem'].notna().sum()} 个XLogP值")
print(f"缺失 {df['xlogp_pubchem'].isna().sum()} 个XLogP值")
df
```
